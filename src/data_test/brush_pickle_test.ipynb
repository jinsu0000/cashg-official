```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
with open("../../dataset/eng/char_pickles/0.pkl", "rb") as f:
    writer_data = pickle.load(f)


i = 10
k = 2
first_char = list(writer_data.keys())[i]
sample = writer_data[first_char][k]
traj = sample["trajectory"]
img = sample["image"]


x = traj[:, 0]
y = traj[:, 1]
eos = traj[:, 2]


print(f"Character: {first_char}")
print("x[:10] =", x[:10])
print("y[:10] =", y[:10])
print("eos[:10] =", eos[:10])


plt.figure(figsize=(12, 4))


plt.subplot(1, 3, 1)
plt.plot(x, y, marker='o')
plt.title("Trajectory (x vs y)")
plt.gca().invert_yaxis()


plt.subplot(1, 3, 2)
colors = ['red' if e == 0 else 'blue' for e in eos]
plt.scatter(x, y, c=colors, s=10)
plt.title("x/y with EOS (red=pen, blue=eos)")
plt.gca().invert_yaxis()


plt.subplot(1, 3, 3)
plt.imshow(img, cmap='gray')
plt.title("Sample Image")

plt.tight_layout()
plt.show()
```
